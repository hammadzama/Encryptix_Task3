```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
credit = pd.read_csv("creditcard.csv")
```

```python
credit.head()
```

```python
credit.info()
```

```python
#checking if there is null value in the data
credit.isnull().sum()
```

```python
# distribution of legit and fraudulent transaction
credit['Class'].value_counts()
```

```python
legit = credit[credit.Class ==0]
fraud = credit[credit.Class ==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_data = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
X=new_data.drop(columns='Class',axis=1)
Y=new_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2,stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

LogisticRegression Model

```python
from sklearn.tree import DecisionTreeClassifier
model_dtc=DecisionTreeClassifier()
model_dtc.fit(X_train,Y_train)
```

```python
print(Y_train)
```

Accuracy Score

```python
X_train_prediction = model_dtc.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data : ', (training_data_accuracy)*100)
```

```python
X_test_prediction = model_dtc.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on testing data : ', (training_data_accuracy)*100)
```
